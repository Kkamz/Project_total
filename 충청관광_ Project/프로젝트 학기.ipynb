### 토픽모델링

```python
import numpy as np
import pandas as pd
import re
import matplotlib.pyplot as plt
from konlpy.tag import Kkma
kkma = Kkma()
import time
import datetime
start = time.time()
from gensim import corpora, models
import gensim
```

```python
topic = pd.read_csv('설문조사(충청도이미지).csv',encoding = 'utf-8')
```

```python
len(topic)
```

```python
topic.head()
```

```python
topic.columns = ['number','image']
```

```python
topic.head()
```

```python
topic_model = list(topic.image)

topic_model[0:2]
```

```python
from string import punctuation
def strip_punctuation(s):
    return ''.join(c for c in s if c not in punctuation)

clean_title = []
for sent in topic_model:
    clean = strip_punctuation(sent)
    clean_title.append(clean)
    
clean_title[0:1]
```

```python
dataset = []
for i in range(len(clean_title)) :
    dataset.append(kkma.nouns(re.sub('[^가-힣a-zA-Z\s]', '', clean_title[i])))

sec = time.time()-start
times = str(datetime.timedelta(seconds = sec)).split(".")
times = times[0]
print(times)
```

```python
clean_title2 = []
for i in range(len(clean_title)) :
    clean_title2.append(re.sub('것',' ',clean_title[i]))
    
clean_title3 = []
for i in range(len(clean_title2)):
    clean_title3.append(re.sub('시티', ' ', clean_title2[i]))

clean_title4 = []
for i in range(len(clean_title3)):
    clean_title4.append(re.sub('뿐', ' ', clean_title3[i]))
    
clean_title5 = []
for i in range(len(clean_title4)):
    clean_title5.append(re.sub('것', ' ', clean_title4[i]))
    
clean_title6 = []
for i in range(len(clean_title5)):
    clean_title6.append(re.sub('대의', ' ', clean_title5[i]))
```

```python
high_score_reviews = dataset
high_score_reviews = [[y for y in x if not len(y) == 1]
                     for x in high_score_reviews]
dictionary = corpora.Dictionary(high_score_reviews)
corpus = [dictionary.doc2bow(text) for text in high_score_reviews]
```

```python
import matplotlib.pyplot as ply
from gensim.models import CoherenceModel

coherence_values=[]
for i in range(2,15):
    ldamodel = gensim.models.ldamodel.LdaModel(corpus, num_topics=i, id2word = dictionary)
    coherence_model_lda = CoherenceModel(model=ldamodel, texts = high_score_reviews, dictionary=dictionary,topn=10)
    coherence_lda = coherence_model_lda.get_coherence()
    coherence_values.append(coherence_lda)
```

```python
x = range(2,15)
plt.plot(x, coherence_values)
plt.xlabel("number of topics")
plt.ylabel("coherence score")
plt.show()
```

```python
perplexity_values = []
for i in range(2,20):
    ldamodel = gensim.models.ldamodel.LdaModel(corpus, num_topics = i, id2word = dictionary)
    perplexity_values.append(ldamodel.log_perplexity(corpus))
```

```python
x=range(2,20)
plt.plot(x, perplexity_values)
plt.xlabel("number of topics")
plt.ylabel("perplexity_values")
plt.show()
```

```python
ldamodel = gensim.models.ldamodel.LdaModel(corpus, num_topics=3, alpha=0.1, id2word = dictionary)
ldamodel.print_topics(num_words=20)
```

```python
import pyLDAvis
import pyLDAvis.gensim

pyLDAvis.enable_notebook()
vis = pyLDAvis.gensim.prepare(ldamodel, corpus, dictionary)
vis
```

### 기상데이터 다중회귀분석

```python
# area : 지점명 
#mean_tem: 평균기온(°C) 
#min_tem : 최저기온(°C) 
#max_tem : 최고기온(°C) 
#mean_mintem : 평균최저기온(°C) 
#mean_maxtem: 평균최고기온(°C) 
# mean_seapre : 평균해면기압(hPa)
#sum_rain : 합계 강수량(mm) 
#daymax_rain : 일최다 강수량(mm) 
#mean_humi : 평균 상대습도(%)
#sum_suntime : 합계 일조시간(hr) 
#per_suntime : 일조율(%) 
#sum_sunstr : 합계 일사량(MJ/m2) 
#mean_wind : 평균 풍속(m/s) 
#max_winddir : 최다 풍향(16방위) 
#per_tourist : 방문객(지역별퍼센트-광역단위)(%)
```

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
wea_2018_ = pd.read_excel("2018 기상데이터(방문자퍼센트추가).xlsx")
wea_2018_.to_csv("2018 기상데이터(방문자퍼센트추가).csv")
wea_2018 = pd.read_csv('2018 기상데이터(방문자퍼센트추가).csv', encoding='UTF-8')
wea_2018.head()
```

```python
wea_2019_ = pd.read_excel("2019 기상데이터(방문자퍼센트추가).xlsx")
wea_2019_.to_csv("2019 기상데이터(방문자퍼센트추가).csv")
wea_2019 = pd.read_csv('2019 기상데이터(방문자퍼센트추가).csv', encoding='UTF-8')
wea_2019.head()
```

```python
nCar = wea_2018.shape[0]
nVar = wea_2018.shape[1]

print(nCar,nVar)

wea_2018.info()
```

```python
nCar = wea_2019.shape[0]
nVar = wea_2019.shape[1]

print(nCar,nVar)

wea_2019.info()
```

```python
variables = wea_2018.columns[:-2].tolist() ## 설명 변수 리스트
 
y = wea_2018['per_tourist'] ## 반응 변수
selected_variables = [] ## 선택된 변수들
sl_enter = 0.05
sl_remove = 0.05
 
sv_per_step = [] ## 각 스텝별로 선택된 변수들
adjusted_r_squared = [] ## 각 스텝별 수정된 결정계수
steps = [] ## 스텝
step = 0
while len(variables) > 0:
    remainder = list(set(variables) - set(selected_variables))
    pval = pd.Series(index=remainder) ## 변수의 p-value
    ## 기존에 포함된 변수와 새로운 변수 하나씩 돌아가면서 
    ## 선형 모형을 적합한다.
    for col in remainder: 
        X = wea_2018[selected_variables+[col]]
        X = sm.add_constant(X)
        model = sm.OLS(y,X).fit()
        pval[col] = model.pvalues[col]
 
    min_pval = pval.min()
    if min_pval < sl_enter: ## 최소 p-value 값이 기준 값보다 작으면 포함
        selected_variables.append(pval.idxmin())
        ## 선택된 변수들에대해서
        ## 어떤 변수를 제거할지 고른다.
        while len(selected_variables) > 0:
            selected_X = wea_2018[selected_variables]
            selected_X = sm.add_constant(selected_X)
            selected_pval = sm.OLS(y,selected_X).fit().pvalues[1:] ## 절편항의 p-value는 뺀다
            max_pval = selected_pval.max()
            if max_pval >= sl_remove: ## 최대 p-value값이 기준값보다 크거나 같으면 제외
                remove_variable = selected_pval.idxmax()
                selected_variables.remove(remove_variable)
            else:
                break
        
        step += 1
        steps.append(step)
        adj_r_squared = sm.OLS(y,sm.add_constant(wea_2018[selected_variables])).fit().rsquared_adj
        adjusted_r_squared.append(adj_r_squared)
        sv_per_step.append(selected_variables.copy())
    else:
        break
```

```python
selected_variables
```

```python
fig = plt.figure(figsize=(10,10))
fig.set_facecolor('white')
 
font_size = 15
plt.xticks(steps,[f'step {s}\n'+'\n'.join(sv_per_step[i]) for i,s in enumerate(steps)], fontsize=12)
plt.plot(steps,adjusted_r_squared, marker='o')
    
plt.ylabel('Adjusted R Squared',fontsize=font_size)
plt.grid(True)
plt.show()
```
